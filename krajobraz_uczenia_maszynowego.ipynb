```python
datapath = Path() /"data"/"lifesat" #def ścieżkę do bieżącego katalogu
# wczytujemy zbiory
oecd_bli=pd.read_csv(datapath/"oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath/"gdp_per_capita.csv")
```

```python
oecd_bli
```

```python
gdp_per_capita
```

```python
#przetwarzamy dane
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename(columns={"GDP per capita, PPP (constant 2017 international $)": "GDP per capita (USD)"})
gdp_per_capita_2020
```

```python
oecd_bli["Indicator"].value_counts()
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(
    index = "Country",
    columns = "Indicator",
    values = "Value"
)
oecd_bli
```

```python
#zresetujemy ...
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
oecd_bli.info()
```

```python
full_country_stats = oecd_bli.merge(
    gdp_per_capita_2020,
    left_on="Country",
    right_on="Entity",
    how="inner"
)
full_country_stats
```

```python
# ograniczamy się do interesujących nas kolumn i sortujemy
full_country_stats = full_country_stats[["Country", "Life satisfaction", "GDP per capita (USD)"]]
full_country_stats = full_country_stats.sort_values(by="GDP per capita (USD)")
full_country_stats
```

```python
full_country_stats.to_csv(
    datapath / "lifesat_full.csv",
    index=False
)
```

```python
min_gdp = 23_500
max_gdp = 62_500
country_stats = full_country_stats[(full_country_stats["GDP per capita (USD)"] >= min_gdp)& (full_country_stats["GDP per capita (USD)"] <= max_gdp)]
country_stats.to_csv(datapath / "lifesat.csv", index=False)
```

```python
lifesat = pd.read_csv("data/lifesat/lifesat.csv")
lifesat.rename(columns={"GDP per capita (USD)": "PKB na mieszkańca (USD)"}, inplace=True)
lifesat.rename(columns={"Life satisfaction": "Satysfakcja z życia"}, inplace=True)
lifesat
```

```python
lifesat.plot(kind='scatter', grid=True, x="PKB na mieszkańca (USD)", y='Satysfakcja z życia')
plt.show()
```

```python
# przygotowuje zmienne w modelu
X= lifesat[["PKB na mieszkańca (USD)"]]
y = lifesat[["Satysfakcja z życia"]]
#wybiera model liniowy
model = LinearRegression()
#trenuje model
model.fit(X,y)
#obliczamy współczynniki modelu
theta_0=model.intercept_[0]
theta_1=model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
```

```python
model.coef_
```

```python
# wizualizujemy dane
lifesat.plot(kind='scatter', grid=True, x="PKB na mieszkańca (USD)", y='Satysfakcja z życia')
plt.plot(X, theta_0 +theta_1*X)
plt.text(45000, 6.5, fr"$\theta_0={theta_0.round(2)}$", color="blue")
plt.text(45000, 6.3, fr"$\theta_1={theta_1*1e5:.2f}\cdot 10^{{-5}}$", color="blue") #.2f - formatuj zapis, zapis z latechu
plt.show()
```

```python
cyprus_gdp_per_capita_2020 = (gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]])
cyprus_gdp_per_capita_2020.rename(columns={"GDP per capita (USD)": "PKB na mieszkańca (USD)"}, inplace=True)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction=model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction
```

```python
# wybiera model regresji 3-najbliższych sąsiadów 
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=3)
# trenuje model_knn
model_knn.fit(X,y)
#uzyskuje przewidywanie dla Cypru
model_knn.predict(cyprus_gdp_per_capita_2020)
```
